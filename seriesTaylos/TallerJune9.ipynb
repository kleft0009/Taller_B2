# Taller 03: series de Taylor y polinomios de Lagrange

Grafique las curvas de las series de Taylor de varios órdenes para los siguientes casos:=0

```python
import sympy as sp

x = sp.Symbol('x')
f = sp.cos(x)

# Orden
ordenes = [0, 1, 2, 3, 4]
           
# Funcion
graficas = [f]
labels = ['cos(x)']

for i in range(5):
    polinomio = sp.series(f, x, 0, n=i+1).removeO()
    print(f"Orden {i}: {polinomio}")

for n in ordenes:
    p = sp.series(f, x, 0, n+1).removeO()
    graficas.append(p)
    labels.append(f'Orden {n}')


p_plot = sp.plot(*graficas, (x, -1.5, 1.5), 
                 xlim=(-1.5, 1.5), 
                 ylim=(-0.5, 1.5),
                 show=False, 
                 title='Serie de Taylor',
                 legend=True)

# Leyenda
for i, label in enumerate(labels):
    p_plot[i].label = label

p_plot.show()
```

Encuentre el polinomio de Lagrange para los siguientes datos y grafique:

```python
import sympy as sp

x = sp.Symbol('x')

x_datos = [0, 30, 60, 90]
y_datos = [0, 0.5, float((sp.sqrt(3)/2).evalf()), 1]

# Polinomio de Lagrange
polinomio = sp.interpolate(list(zip(x_datos, y_datos)), x)

print(polinomio)

graficas = [polinomio]
labels = [f"Polinomio de Lagrange"]

p_plot = sp.plot(
    *graficas,
    (x, -600, 600),          
    xlim=(-650, 650),        
    ylim=(-1000, 1500),      
    show=False,             
    title="Interpolación de Lagrange",
    legend=True
)

# Etiquetas
for i, label in enumerate(labels):
    p_plot[i].label = label

p_plot.show()
```
