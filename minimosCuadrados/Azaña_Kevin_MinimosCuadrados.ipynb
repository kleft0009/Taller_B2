```python
import sympy as sp

def least_squares_poly(data, degree):
    x = sp.symbols('x')
    coeffs = sp.symbols(f'a0:{degree+1}')
    model = sum(coeffs[i] * x**i for i in range(degree + 1))

    residuals = [model.subs(x, xi) - yi for xi, yi in data]
    S = sum(r**2 for r in residuals)

    equations = [sp.diff(S, c) for c in coeffs]
    solution = sp.solve(equations, coeffs, dict=True)

    return model, solution

# Ejemplo
data_points = [(1, 2), (2, 3), (3, 5)]
model, solution = least_squares_poly(data_points, degree=1)

print("Modelo:", model)
print("Solución:", solution)
```
