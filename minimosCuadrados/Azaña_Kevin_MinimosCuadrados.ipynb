```python
import numpy as np
import matplotlib.pyplot as plt

xs = ([
    3.38, 0.35, 2.07, -0.45, -0.55,
    -1.06, -2.77, 3.08, -3.98, -5.0,
    -3.18, -1.96, 2.37, -1.66, 4.09
])

ys = ([
    -0.04, 1.28, 0.65, 2.18, 1.70,
    1.96, 3.36, 0.18, 3.35, 4.16,
    2.95, 2.34, 0.38, 1.75, -1.06
])

def lineal(xs, ys):

    xs = np.array(xs, dtype=float)
    ys = np.array(ys, dtype=float)

    sum_y = np.sum(ys)
    sum_x = np.sum(xs)

    n = len(xs)

    sum_xy = 0
    sum_x2 = 0

    for x, y in zip(xs, ys):
        sum_xy += x * y
        sum_x2 += x**2

    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)

    b = (sum_y - m * sum_x) / n

    y_pred = m * xs + b

    errores_abs = np.abs(ys - y_pred)

    error_promedio = np.mean(errores_abs)
    desviacion = np.std(errores_abs)

    idx_max = np.argmax(errores_abs)
    idx_min = np.argmin(errores_abs)

    print(f"y = {m:.4f}x + {b:.4f}")
    print("Error absoluto promedio:", error_promedio)
    print("Desviación estándar:", desviacion)

    print("\nError máximo")
    print(xs[idx_max], ys[idx_max])

    print("\nError mínimo")
    print(xs[idx_min], ys[idx_min])

    plt.figure(figsize=(8,5))

    plt.scatter(xs, ys, label="Datos")
    plt.plot(xs, y_pred, label="Lineal")

    plt.scatter(xs[idx_max], ys[idx_max],
                s=150,
                marker='x',
                color='red',
                label='Error máximo')

    plt.scatter(xs[idx_min], ys[idx_min],
                s=150,
                marker='o',
                color='green',
                label='Error mínimo')

    plt.legend()
    plt.show()

print(lineal(xs, ys))
```

```python
import numpy as np

xs = ([
    3.38, 0.35, 2.07, -0.45, -0.55,
    -1.06, -2.77, 3.08, -3.98, -5.0,
    -3.18, -1.96, 2.37, -1.66, 4.09
])

ys = ([
    -0.04, 1.28, 0.65, 2.18, 1.70,
    1.96, 3.36, 0.18, 3.35, 4.16,
    2.95, 2.34, 0.38, 1.75, -1.06
])

def cuadrada(xs, ys):

    xs = np.array(xs, dtype=float)
    ys = np.array(ys, dtype=float)

    n = len(xs)

    sx = np.sum(xs)
    sy = np.sum(ys)

    sx2 = sx3 = sx4 = 0
    sxy = sx2y = 0

    for x, y in zip(xs, ys):
        sx2 += x**2
        sx3 += x**3
        sx4 += x**4
        sxy += x*y
        sx2y += (x**2)*y

    A = np.array([
        [sx2, sx, n],
        [sx3, sx2, sx],
        [sx4, sx3, sx2]
    ])

    B = np.array([
        sy,
        sxy,
        sx2y
    ])

    a, b, c = np.linalg.solve(A, B)

    y_pred = a*xs**2 + b*xs + c

    errores_abs = np.abs(ys - y_pred)

    error_promedio = np.mean(errores_abs)
    desviacion = np.std(errores_abs)

    idx_max = np.argmax(errores_abs)
    idx_min = np.argmin(errores_abs)

    print(f"y = {a:.4f}x² + {b:.4f}x + {c:.4f}")

    print("Error absoluto promedio:", error_promedio)
    print("Desviación estándar:", desviacion)

    print("\nError máximo")
    print(xs[idx_max], ys[idx_max])

    print("\nError mínimo")
    print(xs[idx_min], ys[idx_min])

    x_line = np.linspace(min(xs), max(xs), 300)
    y_line = a*x_line**2 + b*x_line + c

    plt.figure(figsize=(8,5))

    plt.scatter(xs, ys, label="Datos")
    plt.plot(x_line, y_line, label="Cuadrática")

    plt.scatter(xs[idx_max], ys[idx_max],
                s=150,
                marker='x',
                color='red',
                label='Error máximo')

    plt.scatter(xs[idx_min], ys[idx_min],
                s=150,
                marker='o',
                color='green',
                label='Error mínimo')

    plt.legend()
    plt.show()

print(cuadrada(xs,ys))
```

```python
import numpy as np

xs = ([
    3.38, 0.35, 2.07, -0.45, -0.55,
    -1.06, -2.77, 3.08, -3.98, -5.0,
    -3.18, -1.96, 2.37, -1.66, 4.09
])

ys = ([
    -0.04, 1.28, 0.65, 2.18, 1.70,
    1.96, 3.36, 0.18, 3.35, 4.16,
    2.95, 2.34, 0.38, 1.75, -1.06
])

def cubica(xs, ys):

    xs = np.array(xs, dtype=float)
    ys = np.array(ys, dtype=float)

    n = len(xs)

    sx = np.sum(xs)
    sy = np.sum(ys)

    sx2 = sx3 = sx4 = sx5 = sx6 = 0
    sxy = sx2y = sx3y = 0

    for x, y in zip(xs, ys):

        sx2 += x**2
        sx3 += x**3
        sx4 += x**4
        sx5 += x**5
        sx6 += x**6

        sxy += x*y
        sx2y += (x**2)*y
        sx3y += (x**3)*y

    A = np.array([
        [sx3, sx2, sx, n],
        [sx4, sx3, sx2, sx],
        [sx5, sx4, sx3, sx2],
        [sx6, sx5, sx4, sx3]
    ])

    B = np.array([
        sy,
        sxy,
        sx2y,
        sx3y
    ])

    a, b, c, d = np.linalg.solve(A, B)

    y_pred = a*xs**3 + b*xs**2 + c*xs + d

    errores_abs = np.abs(ys - y_pred)

    error_promedio = np.mean(errores_abs)
    desviacion = np.std(errores_abs)

    idx_max = np.argmax(errores_abs)
    idx_min = np.argmin(errores_abs)

    print(f"y = {a:.4f}x³ + {b:.4f}x² + {c:.4f}x + {d:.4f}")

    print("Error absoluto promedio:", error_promedio)
    print("Desviación estándar:", desviacion)

    print("\nError máximo")
    print(xs[idx_max], ys[idx_max])

    print("\nError mínimo")
    print(xs[idx_min], ys[idx_min])

    x_line = np.linspace(min(xs), max(xs), 300)
    y_line = a*x_line**3 + b*x_line**2 + c*x_line + d

    plt.figure(figsize=(8,5))

    plt.scatter(xs, ys, label="Datos")
    plt.plot(x_line, y_line, label="Cúbica")

    plt.scatter(xs[idx_max], ys[idx_max],
                s=150,
                marker='x',
                color='red',
                label='Error máximo')

    plt.scatter(xs[idx_min], ys[idx_min],
                s=150,
                marker='o',
                color='green',
                label='Error mínimo')

    plt.legend()
    plt.show()

print(cubica(xs,ys))
```
